```python
import math
f1 = open(r'one.txt')
list = []
list_out = []
for line in f1:
    c=line[0]
    list.append(c)
print(list)
f1.close()
#####################################
f2 = open(r'one_out.txt')
for line in f2:
    g=line[0]
    list_out.append(g)
print(list_out)
f2.close()
#####################################
tn=0
fn=0
tp=0
fp=0
for cun in range(len(list)):
    if(list[cun]=='1' and list_out[cun]=='1'):#TP
        tp=tp+1
    elif(list[cun]=='1' and list_out[cun]=='0'):#FP
        fp=fp+1
    elif(list[cun]=='0' and list_out[cun]=='0'):#TN
        tn=tn+1
    elif(list[cun]=='0' and list_out[cun]=='1'):#FN
        fn=fn+1

sensitivity=tp/(tp+fn) #靈敏度
precision=tp/(tp+fp) #精確率
specificity=tn/(tn+fp) #專一性
accuracy=(tp+tn)/(tp+tn+fp+fn) #準確率
F_score=2*precision*sensitivity/(precision+sensitivity) #F-score
MCC=(tp*tn-fp*fn)/math.sqrt((tp+fp)*(tp+fn)*(tn+fp)*(tn+fn)) #Matthews Correlation Coefficient

print("TP:",tp)
print("FP:",fp)
print("TN:",tn)
print("FN:",fn)

print("sensitivity:",sensitivity)
print("precision:  ",precision)
print("specificity:",specificity)
print("accuracy:   ",accuracy)
print("F-score:    ",F_score)
print("MCC:        ",MCC)
```
